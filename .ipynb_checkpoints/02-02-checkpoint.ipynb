```python
import numpy as np
import pandas as pd
from numpy.random import randn
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

```python
x = range(1,10)
y = [1,2,3,4,0,7,2,5,9]
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1,9])
ax.set_ylim([0,10])

ax.set_xticks([1,2,3,4,5,6,7,8])
ax.set_yticks([0,1,2,9,4,5,6])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1,9])
ax.set_ylim([0,10])

ax.set_xticks([1,2,3,4,5,6,7,8])
ax.set_yticks([0,1,2,9,4,5,6])

ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```
